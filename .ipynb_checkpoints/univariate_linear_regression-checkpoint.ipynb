```python
from typing import List

class LinearRegression:
    """
    Univariate implementation of the Linear Regression model
    using the Mean Squared Error loss function and gradient descent
    """

    def __init__(self, w: float = 1.0, b: float = 0.0):
        self.w = w
        self.b = b

    def predict(self, x_predict: List[float]) -> List[float]:
        return [x * self.w + self.b for x in x_predict]

    @staticmethod
    def mean_squared_error(y_pred: List[float], y_actual: List[float], m: float) -> float:
        return (1 / 2 * m) * sum((y_pred[i] - y_actual[i]) ** 2 for i in range(m))

    @staticmethod
    def compute_w_gradient(x_train: List[float], errors: List[float], m: float) -> float:
        """
        Computes the gradient for the parameter w.
        It's computed from the partial derivative of the MSE loss function J
        with respect to w
        """
        return (1 / m) * sum(errors[i] * x_train[i] for i in range(m))

    @staticmethod
    def compute_b_gradient(errors: List[float], m: float) -> float:
        """
        Computes the gradient for the parameter b.
        It's computed from the partial derivative of the MSE loss function J
        with respect to b
        """
        return (1 / m) * sum(errors)

    def fit(self, x_train: List[float], y_train: List[float], learning_rate: float = 0.01, epochs: int = 10):
        m = len(x_train)
        for epoch in range(epochs):
            y_pred = self.predict(x_train)
            errors = [y_pred[i] - y_train[i] for i in range(m)]
            self.w -= learning_rate * self.compute_w_gradient(x_train, errors, m)
            self.b -= learning_rate * self.compute_b_gradient(errors, m)
```

```python
import numpy as np

np.random.seed(42)
x_train = np.linspace(0, 10, 20)
y_train = 2 * x_train + 5 + np.random.normal(0, 1, 20)  # y = 2x + 5 + noise

# Generate test data
x_test = np.linspace(0, 15, 100)

model = LinearRegression()
model.fit(x_train.tolist(), y_train.tolist(), epochs=1000)

y_pred = model.predict(x_test.tolist())
```

```python
import matplotlib.pyplot as plt

plt.scatter(x_train, y_train, color='blue', label='Training Data', s=10)
plt.plot(x_test, y_pred, color='red', label='Predictions', linewidth=1)
plt.xlabel('X', fontsize=14)
plt.ylabel('Y', fontsize=14)
plt.legend(fontsize=10)
```
