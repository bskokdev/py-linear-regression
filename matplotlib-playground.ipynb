```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)

n = 100

x1 = np.random.normal(loc=3, scale=1, size=n)
y1 = np.random.normal(loc=3, scale=1, size=n)

x2 = np.random.normal(loc=8, scale=1, size=n)
y2 = np.random.normal(loc=8, scale=1, size=n)

fig, ax = plt.subplots()

ax.scatter(x1, y1, label='Cluster A', marker='x')
ax.scatter(x2, y2, label='Cluster B', marker='o')

ax.plot((12,0), (0,10), 'r-')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.legend()

plt.show()
```
